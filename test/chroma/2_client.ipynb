```python
import chromadb
```

```python
#1.创建客户端
#1.1创建临时客户端
# client=chromadb.EphemeralClient()
#1.2持久化客户端
# client=chromadb.PersistentClient(path='./chroma_db')
#1.3网络客户端
client=chromadb.HttpClient(port=10240,host='127.0.0.1')
```

```python
#2.创建collection
collection=client.get_collection('test_collection')
```

```python
collection=client.create_collection('test_collection')
```

```python
#列出所有的collection
client.list_collections()
```

```python
#3.插入数据
collection.add(
    ids=['id1','id2','id3'],
    documents=[
        '1111',
        '2222',
        '3333',
    ]
)
```

```python
#4.查询
collection.query(
    ids=['id2'],
)
```
