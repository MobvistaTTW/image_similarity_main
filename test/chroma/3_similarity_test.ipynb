```python
import chromadb
```

```python
#1.创建客户端
client=chromadb.Client()
```

```python
#2.创建集合
# collection=client.create_collection(name='my_collection')
#删除集合
client.delete_collection(name='my_collection')
collection=client.create_collection(
    name='my_collection',
    configuration={
        'hnsw':{
            'space':'cosine'
        }
        #ip,l2,cosine;通常用cosine
    }
)
```

```python
client.list_collections()
```

```python
collection.peek()
```

```python
#3.插入数据
collection.add(
    ids=['id1','id2','id3'],
    documents=[
        '1111',
        '2222',
        '3333',
    ],
    embeddings=[
        [1.0,2,3],
        [4,5,6],
        [7,8,9]
    ]
)
```

```python
collection.peek()
```

```python
#4.查询
collection.query(query_embeddings=[1,2,3])
```
