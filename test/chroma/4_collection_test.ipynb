```python
import chromadb
```

```python
#1.创建客户端
client=chromadb.PersistentClient(path='./chroma_db')
```

```python
#2.获取或创建
collection1=client.create_collection('collection')
collection2=client.get_or_create_collection('collection2')
```

## 查看集合

```python
# 1.列出所有集合
client.list_collections()
```

```python
#2.复制一个集合(他只在内存中,不在数据库中)
collection2_cpy=client.get_collection(name='collection2')
```

```python
print(collection2_cpy)
collection2_cpy.modify(name='collection2_cpy')
print(collection2_cpy)
print(collection2_cpy.id)
print(collection2.id)
print(collection2.id==collection2_cpy.id)
print(client.get_collection('collection2'))#NotFoundError: Collection [collection2] does not exist
```

```python
client.list_collections()
```

## 向集合中添加数据

```python
collection2.add(
    ids=['id1','id2','id3'],
    documents=[
        '1111',
        '2222',
        '3333',
    ],
    embeddings=[
        [1.0,2,3],
        [4,5,6],
        [7,8,9]
    ]
)
```

```python
collection2.peek()
```

```python
print(collection2)
```

```python
client.list_collections()
```

```python
collection2.modify(name='collection2')
client.list_collections()
```

```python
print(collection2_cpy)
print(collection2)
print(collection2.id==collection2_cpy.id)
```

```python
collection2.add(
    ids='id_collection2',
    documents='id_collection2',
    embeddings=[[6,3,1]]
)
collection2_cpy.add(
    ids='id_collection2_cpy',
    documents='id_collection2_cpy',
    embeddings=[[1,5,2]]
                    )
print(collection2.peek())
print(collection2_cpy.peek())
```

## 删除集合
