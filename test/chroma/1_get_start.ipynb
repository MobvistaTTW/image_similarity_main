```python
import chromadb
```

```python
#1.创建chroma客户端
client=chromadb.Client()
```

```python
#2.创建collection
collection=client.get_or_create_collection(
    name='test_collection'
)
```

```python
#3.插入数据
collection.add(
    ids=['4','5','6'],
    documents=[
        "我闻琵琶已叹息",
        "凄凄惨惨戚戚",
        "断肠人在天涯",
    ]
)
```

```python
results=collection.peek()
print(results['embeddings'].shape)
print(results['embeddings'])
print(results)
```

```python
#5.获取数据
collection.get(ids=['2'])
```

```python
#6.根据相似度插询
collection.query(
    query_texts="This is the first document",
    n_results=9
)
```

```python
# 查询相似文档
results = collection.query(
    query_texts="This is the first document",
    n_results=6
)

# 假设相似度得分存储在 results['distances'] 中（需确认实际字段名）
threshold = 0.8  # 设置相似度阈值
filtered_results = []

for i, distance in enumerate(results['distances'][0]):  # 遍历距离列表
    if distance < threshold:  # 距离越小，相似度越高
        filtered_results.append({
            'id': results['ids'][0][i],
            'document': results['documents'][0][i],
            'distance': distance
        })

print(filtered_results)
```
